# ANALYSING CREDIT CARD DATA FROM Deutsche Mark.

```python
import numpy as np
import pandas as pd
from scipy.io import arff
```

```python
    # Loading the data set https://openml.org/search?type=data&sort=runs&status=active&id=31

data = arff.loadarff('../datasets/dataset_31_credit-g.arff')
```

```python
train= pd.DataFrame(data[0])
train.info()
```

```python

train.describe()
```

```python
import matplotlib.pyplot as plt
train.hist(bins=50, figsize=(12, 8))
```

```python
train.plot(kind='scatter', x='age', y='existing_credits', grid=True)
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(train['age'], train['existing_credits'])
print(f'MSE : {mse} || RMSE : {np.sqrt(mse)}')
```
